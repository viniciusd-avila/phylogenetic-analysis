```python
example = bioinfo.gen_df2(df['Sequence'][2],3)
```

```python
from sklearn.preprocessing import StandardScaler
x = example.loc[:, example.columns != 'Fragment'].values
y = example.loc[:, example.columns == 'Fragment'].values
x = StandardScaler().fit_transform(x)
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=2)

principalComponents = pca.fit_transform(x)

principalDf = pd.DataFrame(data = principalComponents
             , columns = ['principal component 1', 'principal component 2'])
```

```python
finalDf = pd.concat([principalDf, example[['Fragment']]], axis = 1)
```

```python
finalDf
```

```python
fig = plt.figure(figsize = (8,8))
ax = fig.add_subplot(1,1,1) 
ax.set_xlabel('Principal Component 1', fontsize = 15)
ax.set_ylabel('Principal Component 2', fontsize = 15)
ax.set_title('2 component PCA', fontsize = 20)

ax.scatter(finalDf.loc[:, 'principal component 1']
               , finalDf.loc[:, 'principal component 2'])
ax.grid()
```
